```python
import os
import conda

conda_file_dir = conda.__file__
conda_dir = conda_file_dir.split('lib')[0]
proj_lib = os.path.join(os.path.join(conda_dir, 'share'), 'proj')
os.environ["PROJ_LIB"] = proj_lib

from mpl_toolkits.basemap import Basemap
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
file = 'selatmalaka.nc4'
ds = Dataset(file)
print(ds)
```

```python
surf = ds.variables['surf_el'][0,:,:]
lats = ds.variables['lat'][:]
lons = ds.variables['lon'][:]
time = ds.variables['time'][:]
ds.close()
```

```python
lon_0 = lons.mean()
lat_0 = lats.mean()
m = Basemap(projection='mill',lat_ts=10,
            llcrnrlon=95.5, urcrnrlon=105,
            llcrnrlat=0.5,urcrnrlat=5.5,
            resolution='h')
Lon, Lat = np.meshgrid(lons,lats)
x, y = m(Lon, Lat)
```

```python
fig = plt.figure(figsize=(12,9))
m.drawcountries()
m.drawcoastlines()
m.drawmapboundary()
m.contourf(x,y,surf,cmap=plt.cm.jet)
b = m.colorbar()
b.set_label('Sea Surface Height',rotation=90)

meridians = np.arange(Lon.min(),Lon.max(),5)
parallels = np.arange(Lat.min(),Lat.max(),5)
plt.title('SSH',fontsize=20)
m.drawparallels(np.arange(0.5, 5.5, 1), labels = [1, 0, 0, 0], linewidth =0.5, color = 'k')
m.drawmeridians(np.arange(95.5, 105, 1), labels = [0, 0, 0, 1], linewidth =0.5, color = 'k')
```
